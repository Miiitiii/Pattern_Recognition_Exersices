```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
# reading csv files
data =  pd.read_csv('D:/HomeWork/iris.data', sep=",",header = None)
```

```python
#Feature 1 and 4 with Transformation
T1 = np.array([0.1 , -1.5])
setosa = data[data[4] == 'Iris-setosa']
virginica = data[data[4] == 'Iris-virginica']
versicolor = data[data[4] == 'Iris-versicolor']

setosa = setosa.loc[: , [0,3]]
virginica = virginica.loc[:, [0,3]]
versicolor = versicolor.loc[:, [0,3]]

setosa =np.matmul(setosa , T1)
virginica =np.matmul(virginica , T1)
versicolor =np.matmul(versicolor , T1)

plt.hist(setosa , color = 'red', label='setosa', edgecolor='None', alpha = 0.7)
plt.hist(virginica , color = 'blue', label='virginica', edgecolor='None', alpha = 0.7)
plt.hist(versicolor , color = 'black', label='versicolor', edgecolor='None', alpha = 0.7)
plt.title('Feature 1 and 4')

plt.ylabel('Probability')
plt.xlabel('Data');
plt.legend()

plt.show()
```
