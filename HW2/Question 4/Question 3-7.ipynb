```python
likelihood = []

for i in range(1,21):
    dist = [binom.pmf(r, i, 0.2) for r in dataSet]
    li = 1
    for j in data :
        li*= dist[j]
    likelihood.append(li)


plt.plot(list(range(1,21)), likelihood , color ='r')
plt.show()
```

```python
likelihood_log = []

for i in range(1,21):
    dist = [binom.pmf(r, i, 0.2) for r in dataSet]
    li = 1
    for j in data :
        li += np.log(dist[j])
    likelihood_log.append(li)


plt.plot(list(range(1,21)), likelihood_log , color ='r')
plt.show()
```
