```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = [0,2,0,3,1,2,2,2,0,2,3,1,0,0,2,3,3,2,3,0]
dataSet = list(set(data))
```

```python
height = []
for num in dataSet:
    height.append(data.count(num))
```

```python
plt.bar(dataSet, height , color ='r')  
plt.title("Bar Chart")
plt.show()
```

```python
my_array = np.array(height)

#Normalise X so the values lie between -1 and 1.
mu = my_array.mean(0) 
sigma = my_array.std(0) # standard deviation: max(x)-min(x)
xn = (my_array - mu) / sigma
```

```python
plt.bar(dataSet, xn , color ='r')  
plt.title("Bar Chart")
plt.show()

xn
```

```python
plt.bar(dataSet, xn, color = "Red")   
plt.plot(dataSet, xn,color="Blue")    
plt.show()
```
