```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import scipy.io as sio
from sklearn.metrics import accuracy_score
import random


# reading csv files
rand = sio.loadmat(r'D:\HomeWork\Pattern\inputs\P3\rand.mat')

points = rand['dataset'][0][0][0]
labels = rand['dataset'][0][0][1]
```

```python
#f for class 0 and s for class 1
def kmeanscluster(X , k , mu , tol , maxIter):
    iteration = 0
    SSE = 100
    finalCenters = 0
    
    while iteration < maxIter and tol < SSE:
        
        newSSE = 0
        
        
        newDatas = list()
        
        for item in range(k):
            dist = []
            for i in range(len(X)):
                dist.append(np.linalg.norm(X[i] - mu[:, item:item+1].reshape(len(mu))))
            newDatas.append(np.array(dist))
            
        
        newLabels = np.argmin(np.array(newDatas), axis=0)
        
        finalCenters = mu
        
        means = list()
        for item in range(k):
            arrangedData = []
            for i in range(len(X)):
                if newLabels[i] == item:
                    arrangedData.append(X[i])
            data = np.array(arrangedData)
            newSSE = newSSE + np.sum((data - mu[:, item:item+1].reshape(len(mu)))**2)
            meanData = data.mean(axis = 0)
            means.append(meanData)
        
        meansArray = []

        for i in range(len(means)-1, -1, -1):
            meansArray = [y for x in [meansArray, means[i]] for y in x]
        mu = np.reshape(meansArray, (len(means[0]), k), order='F')
        
        iteration = iteration + 1
        SSE = newSSE
        
        
    return newLabels , SSE, finalCenters
```

```python
#Question C
#f for class 0 and s for class 1
def kmedoidscluster(X , k , mu , tol , maxIter):
    iteration = 0
    SSE = 100
    finalCenters = 0
    
    while iteration < maxIter and tol < SSE:
        
        newSSE = 0
        
        
        newDatas = list()
        
        for item in range(k):
            dist = []
            for i in range(len(X)):
                dist.append(np.linalg.norm(X[i] - mu[:, item:item+1].reshape(len(mu))))
            newDatas.append(np.array(dist))
            
        
        newLabels = np.argmin(np.array(newDatas), axis=0)
        
        finalCenters = mu
        
        means = list()
        for item in range(k):
            arrangedData = []
            for i in range(len(X)):
                if newLabels[i] == item:
                    arrangedData.append(X[i])
            data = np.array(arrangedData)
            newSSE = newSSE + np.sum((data - mu[:, item:item+1].reshape(len(mu)))**2)
            meanData = data.mean(axis = 0)
            dist1 = []
            for j in range(len(arrangedData)):
                dist1.append(np.linalg.norm(arrangedData[j] - meanData))
            means.append(arrangedData[np.argmin(dist1)])
        
        meansArray = []

        for i in range(len(means)-1, -1, -1):
            meansArray = [y for x in [meansArray, means[i]] for y in x]
        mu = np.reshape(meansArray, (len(means[0]), k), order='F')
        
        SSE = newSSE
        
        iteration = iteration + 1
                
    return newLabels , SSE , finalCenters
```
