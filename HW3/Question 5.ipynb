```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KernelDensity


data = pd.read_csv(r"D:\HomeWork\Pattern\SPR_HW3\inputs\P5\digits.csv")

data.head()
```

```python
#part A
```

```python
pca = PCA(n_components=15)
newDataa = pca.fit_transform(data)


newDataFrame = pd.DataFrame(data = newDataa)


eigenValues = pca.components_
eigenValues
```

```python
#part B
```

```python
params = {"bandwidth": np.logspace(0 , 10 , 20)}

grid = GridSearchCV(KernelDensity(), params)

grid.fit(newDataFrame)
```

```python
print(grid.best_estimator_)
```
